```python
%matplotlib inline
%reload_ext autoreload
%autoreload
SEED = 1234
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from IPython.display import display, HTML
from time import time
import numpy as np
import numpy.ma as ma
np.set_printoptions(precision=4)
np.random.seed(SEED)
import pandas as  pd
#pd.options.display.max_rows
#pd.set_option('display.max_colwidth', -1)
#pd.set_option('display.max_rows', 5)
pd.set_option('display.max_columns', 44)
pd.set_option('precision', 3)
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(50,65), 'figure.dpi':100})
import seaborn as sns
```

```python
obras = pd.read_csv("obras.csv")
```

```python
obras['Numero'] = obras['Numero'].astype("Int64")
obras['Comuna'] = obras['Comuna'].astype("Int64")
obras.info()
#df_datos = df_categoricas[["Car_desc"]].groupby(["Car_desc"])["Car_desc"].count().reset_index(name='Cantidad')
#df_datos["Porcentaje"] = df_datos["Cantidad"]/df_datos["Cantidad"].sum() * 100
#df_datos = df_datos.sort_values(by="Cantidad", ascending = False)
#display(df_datos)
```

```python
barrio = obras[["Barrio"]].fillna('DESCONOCIDO').groupby(["Barrio"])["Barrio"].count().reset_index(name='Cantidad')
barrio = barrio.sort_values(by="Cantidad", ascending = False)
barrio["Porcentaje"] = barrio["Cantidad"]/barrio["Cantidad"].sum() * 100
barrio = barrio.sort_values(by="Cantidad", ascending = False).reset_index()
del barrio['index']
barrio.to_csv('analisisUnivariado/barrio.csv', index = False)

barrio.plot.barh(x='Barrio', y= 'Cantidad', rot=0, figsize=(15,15), fontsize = 13)
plt.savefig('analisisUnivariado/barrio.png') 
display(barrio.head(10))
obras["Barrio"].mode()
```

```python
expedientes = obras[["Expediente"]].fillna('DESCONOCIDO').groupby(["Expediente"])["Expediente"].count().reset_index(name='Cantidad')
expedientes = expedientes.sort_values(by="Cantidad", ascending = False)
expedientes["Porcentaje"] = expedientes["Cantidad"]/expedientes["Cantidad"].sum() * 100
expedientes = expedientes.sort_values(by="Cantidad", ascending = False).reset_index()
del expedientes['index']
expedientes.to_csv('analisisUnivariado/expedientes.csv', index = False)

expedientes.head(15).plot.barh(x='Expediente', y= 'Cantidad', rot=0, figsize=(15,15), fontsize = 13)
plt.savefig('analisisUnivariado/expedientes.png') 
display(expedientes.head(15))

```
